Data types are correctly assigned. However, for the Machine Learning part we want to transform the following three features into numericals: *GraduateOrNot*, *FrequentFlyer*, *EverTravelledAbroad*.


```python
df["GraduateOrNot"].value_counts()
```

```python
df["FrequentFlyer"].value_counts()
```

We transform all "No" answers into 0 and "Yes" into 1, which alligns with the notation of the features *ChronicDiseases* and *TravelInsurance*. This method is also called **Binary encoding**:

```python
columns_to_change = ["GraduateOrNot", "FrequentFlyer", "EverTravelledAbroad"]

df[columns_to_change] = df[columns_to_change].replace({"Yes": 1, "No": 0})

df.head()
```

The last categorial feature *Employment Type* we will transform as well:

```python
df["Employment Type"].value_counts()
```

There are two types of *Employment type*, but without a natural order between the categories (unlike between yes and no). Therefore we'd use **One Hot Encoding**, which creates a separate binary column (*dummy variables*)for each category, ensuring no implicit relationship is assumed. 

However, this ultimately results in the problem that the two columns are perfectly correlated. Avoiding Multicollinearity in the Model is approached by dropping one dummy variable and identifying the remaining one as the reference category. The coefficients for the remaining dummy variables represent the effect of those categories relative to the dropped category.

Consequently, we can simply recode the existing `Employment Type` column as done before with **Binary Encoding** where "Government Sector" becomes 0 and "Private Sector/Self Employed" becomes 1:

```python
columns_to_change = ["Employment Type"]

df[columns_to_change] = df[columns_to_change].replace(
    {"Private Sector/Self Employed": 1, "Government Sector": 0}
)
```

### 6.3. <a id='toc6_3_'></a>[Outliers: identification and handling](#toc0_)

Identifying outliers in the data follows a three step plan:

2) Using a statistical method to calculate the outlier data points
3) Applying a statistical method to flag, drop or transform the outliers

Moreover, a two-level approach is applied to check both types of outliers, *natural* and *anomalous*: the rule-based method and the statistical approach. 

#### 6.3.1. <a id='toc6_3_1_'></a>[Rule-based method](#toc0_)

The rule-based method to indicate outliers is done simultaneously with the statistical outliers approach (see 3.2.2.). The boxplots provide a clear picture if datapoints are breaking the particular rule applied to a feature (e.g. unreasonable years of age). ## 7. <a id='toc7_'></a>[Exploratory Data Analysis (EDA)](#toc0_)

### 7.1. <a id='toc7_1_'></a>[Correlations between predictor variables](#toc0_)

**Blind Analysis**

We intentionally "blind" ourselves to the outcome variable during this first step of the EDA, exploring predictors and distributions without knowing their impact on the dependent variable.


#### 7.1.1. <a id='toc7_1_1_'></a>[Correlation matrix](#toc0_)

```python
from custom_modules.stat_calculations import custom_correlation_with_pvalues

correlation_matrix, p_values_matrix = custom_correlation_with_pvalues(
    df.drop(columns=["TravelInsurance"]), feature_types
)
np.fill_diagonal(correlation_matrix.values, np.nan)
np.fill_diagonal(p_values_matrix.values, np.nan)

mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

fig, ax = plt.subplots(figsize=(10, 6))
cbar_kws_corr = {
    "label": "Pearson/Phi/Spearman Correlation",
    "orientation": "vertical",
    "shrink": 0.7,
    "pad": 0.05,
}

cbar_kws_pval = {
    "label": "P-value",
    "orientation": "vertical",
    "shrink": 0.7,
    "pad": 0.05,
}
annot_kws = {"size": 9}
sns.heatmap(
    p_values_matrix,
    mask=~mask,
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    cbar=True,
    cbar_kws=cbar_kws_pval,
    annot_kws=annot_kws,
    ax=ax,
)
sns.heatmap(
    correlation_matrix,
    mask=mask,
    cmap="BrBG",
    annot=True,
    fmt=".2f",
    cbar=True,
    cbar_kws=cbar_kws_corr,
    annot_kws=annot_kws,
    ax=ax,
)

plt.title(
    "Combined Matrix Heatmap: Data type dependend Correlation and P-Values", pad=15
)
plt.xticks(rotation=50, fontsize=8)
plt.yticks(fontsize=8)
plt.show()
```

The ***Combined Matrix Heatmap*** above shows the relationship between all possible predictors with each other. In the left bottom triangle are the correlation coefficients displayed, and the upper right triangle are the corresponding p-values. 

The correlation coefficients are calculated with the appropriate approach depending on the data types of the two features analysed: The ***Pearson coefficient*** for the correlation between two continuous features (e.g. Age and AnnualIncome), the ***Phi coefficient*** (and chi-squared test p-value) for the correlation between two nominal binary features (e.g. Employment Type and GraduateOrNot), and the ***Spearman coefficient*** for ordinal/binary vs continuous or ordinal vs ordinal correlation (e.g. FamilyMembers and FrequentFlyer).

Collinearity exists, when two or more predictor variables are closely related to one another. Collinearity reduces the accuracy of the estimates of the regression coefficients, causes the standard error to grow and the t-statistic to decrease, thus the *power* of the hypothesis test is reduced, too. The correlation matrix shows the strength of relations between the variables, and the heatmap above highlights them graphically. 

From the heatmap of the correlation matrix, the following moderate **positive correlations** (0.30 to 0.50) can be highlighted:
- `AnnualIncome` + `Employment Type` 
- `AnnualIncome` + `FrequentFlyer`
- `AnnualIncome` + `EverTravelledAbroad`

All of these correlations show also very low p-values (<0.00) and are therefore statistically significant.

### 7.3. <a id='toc7_3_'></a>[First check for multicollinearity](#toc0_)

While ***multicollinearity*** does not hinder a model's predictive capabilities, it poses challenges in ***explanatory modeling*** and impacts the interpretability. This is because the focus lies on understanding the individual contributions of predictors to the outcome variable, which multicollinearity obscures by creating dependencies among explanatory variables. High multicollinearity inflates the standard errors of regression coefficients, complicating statistical inference and hypothesis testing. Consequently, assessing and addressing multicollinearity is a standard practice in explanatory modeling. A broad range of strategies exists for its detection and mitigation, with ***variable selection*** being a common approach to reduce its impact.

But also in the context of Machine Learning and ***predictive modeling*** multicollinearity plays a crucial role and can ultimately result in overfitting. Finally, [multicollinearity can be handled](https://towardsdatascience.com/handling-multi-collinearity-6579eb99fd81) with the following two methods:
1. *Introduce penalization or remove highly correlated variables*: Use lasso and ridge regression to eliminate variables which provide information which is redundant. This can also be achieved by observing the VIF.
2. *Combine highly correlated variables*: Since the collinear variables contain redundant information, combining them into a single variable using methods such as PCA to generate independent variables.


The variance inflation factor (VIF) is used to detect multicollinearity. A VIF above 5 or 10 is commonly used as a guideline to indicate that a feature is highly collinear with other features and may cause multicollinearity issues in a regression or classification model. A VIF greater than 5 suggests that the feature is highly correlated with other variables, and multicollinearity may start to affect the stability of the model. In practice, values greater than 5 indicate that the variable might be contributing redundant information.

### 8.2. <a id='toc8_2_'></a>[Hypotheses formulation](#toc0_)

Based on initial research about Travel Insurances and a first iteration of EDA we formulate the following five hypotheses that should be tested:

1. Higher ***Age*** typically increases awareness of risks, which makes individuals more likely to purchase insurance. 
- Alternative Hypothesis ($H_A(1)$): `Age` has a positive influence on the likelihood of purchasing a travel insurance package $(\beta_{\text{Age}} > 0)$.
- Null Hypothesis ($H_0(1)$): `Age` has no effect on the likelihood of purchasing a travel insurance package $(\beta_{\text{Age}} = 0)$.

2. Higher ***Income*** provides greater financial flexibility, increasing the likelihood of individuals purchasing insurance. 

- Alternative Hypothesis ($H_A(2)$): `Income` has a positive influence on the likelihood of purchasing a travel insurance package $(\beta_{\text{AnnualIncome}} > 0)$.
- Null Hypothesis ($H_0(2)$): `Income` has no effect on the likelihood of purchasing a travel insurance package $(\beta_{\text{AnnualIncome}} = 0)$.

3. A ***larger family*** increases the likelihood of traveling with others, which elevates the associated risks and consequently heightens the need for travel insurance coverage.
- Alternative Hypothesis ($H_A(3)$): `FamilyMembers` has a positive influence on the likelihood of purchasing a travel insurance package $(\beta_{\text{FamilyMembers}} > 0)$.
- Null Hypothesis ($H_0(3)$): `FamilyMembers` has no effect on the likelihood of purchasing a travel insurance package $(\beta_{\text{FamilyMembers}} = 0)$.
 
4. ***Frequent air travel*** raises the risk associated with journeys, thereby increasing the need for travel insurance coverage.
- Alternative Hypothesis ($H_A(4)$): `FrequentFlyer` has a positive influence on the likelihood of purchasing a travel insurance package $(\beta_{\text{FrequentFlyer}} > 0)$.
- Null Hypothesis ($H_0(4)$): `FrequentFlyer` has no effect on the likelihood of purchasing a travel insurance package $(\beta_{\text{FrequentFlyer}} = 0)$.

5. ***Previous international travel experience*** raises the likelihood of frequent future travel abroad. This, in turn, heightens the risk during travel and increases the demand for travel insurance.
- Alternative Hypothesis ($H_A(5)$): `EverTravelledAbroad` has a positive influence on the likelihood of purchasing a travel insurance package $(\beta_{\text{EverTravelledAbroad}} > 0)$.
- Null Hypothesis ($H_0(5)$): `EverTravelledAbroad` has a positive influence on the likelihood of purchasing a travel insurance package $(\beta_{\text{EverTravelledAbroad}} = 0)$.

The significance level set for these hyotheses to test is set a $\alpha$ level of 0.05.

The coefficient test in logistic regression is conceptually similar to performing individual tests for the significance of each predictor in isolation. However, logistic regression modeling provides several advantages:  

- **Control for confounding variables:** Logistic regression evaluates the effect of each predictor variable while accounting for the influence of other predictors in the model.  
- **Quantify the magnitude of the effect:** Logistic regression coefficients represent changes in the log odds of the outcome for a one-unit increase in the predictor. When exponentiated, these coefficients yield odds ratios, making the results interpretable in terms of the relative likelihood of the outcome.  
- **Perform hypothesis testing in a multivariate context:** Coefficient tests (such as Wald z-tests) allow for the direct evaluation of predictor significance within the context of the full model, ensuring that the contribution of each variable is assessed in combination with others.

Therefore, we perform hypothesis testing for logistic regression coefficients in the context of the established statistical model to determine whether each predictor significantly contributes to predicting the target variable.

### 8.9. <a id='toc8_9_'></a>[Model evalution metrics](#toc0_)

When evaluting the ***baseline classifier*** (`EverTravelledAbroad`) as well as the 3 versions of the Logistic Regression model derived via statistical learning, we've seen multiple metrices. Furthermore we want to focus in particular on one: **Recall**.

The reason lies back within the main objective and problem definition (see 2. Problem discovery). Recall measures the ability of the model to correctly identify all the customers who are actually interested in purchasing the insurance, meaning it captures the proportion of true positives (customers who actually want the insurance) out of all actual positives. There are two main reasons why recall matters for our problem context:

1. Avoiding missed opportunities:

- If our primary concern is not missing any potentially interested customers, focusing on recall is critical. In other words, we want to ensure that our model identifies as many customers as possible who might be interested in the travel insurance, even if that means it might occasionally make some false positives (predicting interest when there is none).
- Recall ensures that we catch most of the actual interested customers, minimizing the risk of overlooking potential buyers who could generate revenue.
2. Cost of missing a true positive:

- If failing to target an interested customer (a false negative) is more costly than targeting a few non-interested customers (false positives), recall should be prioritized. For instance, if we're sending marketing emails or ads for the insurance, missing out on customers who would buy the insurance means we lose out on potential sales.
- Increasing recall helps us to reach as many interested customers as possible, reducing the chance of leaving money on the table by not targeting someone who would buy the insurance package.


**Balancing recall and precision**

There is a tradeoff between **precision** and recall. If we increase recall (by, for example, lowering the threshold for predicting a customer as interested), precision may drop, because we'll start to predict more false positives. Precision (the proportion of positive predictions that are actually correct) is crucial we you want to minimize false positives — in other words, situations where the model incorrectly predicts that a customer will be interested in the insurance when they actually will not. False positives are particularly problematic because they could result in wasted marketing efforts, unnecessary follow-ups, and customer frustration.

A good approach is to find a balance between precision and recall based on our business objectives. We can use the **F1-score**, which is the harmonic mean of precision and recall, to evaluate the overall performance when there is an imbalance between the two metrics.

#### 9.2.3. <a id='toc9_2_3_'></a>[Model training and evaluation](#toc0_)

To better understand the generalization performance of our model and maybe find insights on how to improve it, we compare the *testing error* with the *training error*. In fact, the *testing error* does not result from the test data but from the ***validation data***, which is accomplished by **crossvalidation**.

The crossvalidation through the piple handles the fitting process (model training) internally. Thus we can directly derive our scores of each model:

```python
pipelines = [log, log_standard, log_normal]


def model_evaluation(pipelines, data_train, target_train):
    for pipeline in pipelines:
        metrics = ["accuracy", "precision", "recall", "f1"]
        cv_results = cross_validate(
            pipeline,
            data_train,
            target_train,
            cv=10,
            scoring=metrics,
            return_estimator=True,
        )
        cv_results = pd.DataFrame(cv_results)

        accuracy_score = cv_results["test_accuracy"]
        precision_score = cv_results["test_precision"]
        recall_score = cv_results["test_recall"]
        f1_score = cv_results["test_f1"]

        print(
            f"Model: {pipeline} \n"
            f"The mean cross-validation scores are:\n"
            f"  - Accuracy: {accuracy_score.mean():.3f} ± {accuracy_score.std():.3f}\n"
            f"  - Precision: {precision_score.mean():.3f} ± {precision_score.std():.3f}\n"
            f"  - Recall: {recall_score.mean():.3f} ± {recall_score.std():.3f}\n"
            f"  - F1-Score: {f1_score.mean():.3f} ± {f1_score.std():.3f}\n"
        )


model_evaluation(pipelines, data_train, target_train)
```

The crossvalidation results show, that the ***Logistic Regression model using standardized features*** (standardscaler) shows the best average scores across all three key metrics. We use the ***Logistic Regression model using standardized features*** again, which performed best in the previous section.

#### 9.3.2. <a id='toc9_3_2_'></a>[Evaluation and hyperparameter C tuning](#toc0_)

The logistic regression classifier has a **hyperparameter C** which controls the ***regularization strength***. A small C provides a more regularized model, whereas a non-regularized model is obtained with an infinitely large value of C.

We want to let the model search itself for the best hyperparamter C. We use **Grid search** because it exhaustively evaluates all combinations of hyperparameters in the specified range. The range of possible values for C is small and we can afford the computation time to evaluate all combinations. Moreover, we are confident about the range of C values to explore and don't need a large search space.




We set the known pipeline including ***Standardization***:

```python
log_standard = Pipeline(
    [
        ("preprocessor", StandardScaler()),
        (
            "classifier",
            LogisticRegression(max_iter=500, class_weight="balanced"),
        ),
    ]
)
log_standard
```

Then we have to set a range of particular values for the Grid Search to check. The effect of C on model performance is typically nonlinear. Testing logarithmic values ensures us to cover a wide range of possibilities while focusing on meaningful increments:

```python
param_grid = {
    "classifier__C": np.logspace(-3, 2, num=6),
}
```

We use an inner cross-validation for the selection of the hyperparameters (via `GridSearchCV`) and an outer cross-validation for the evaluation of generalization performance of the refitted tuned model (via `cross_validate`). This approach is called ***nested cross-validation***:

```python
log_standard_grid = GridSearchCV(log_standard, param_grid=param_grid, n_jobs=2, cv=10)

cv_results = cross_validate(
    log_standard_grid, data_train, target_train, cv=10, n_jobs=2, return_estimator=True
)
```

We can also show which values of C performed best across the 10 folds best:

```python
best_c_values = [
    estimator_in_fold.best_params_["classifier__C"]
    for estimator_in_fold in cv_results["estimator"]
]

c_counter = Counter(best_c_values)

plt.figure(figsize=(4, 3))
plt.bar(
    [str(c) for c in c_counter.keys()],
    c_counter.values(),
    color="skyblue",
    edgecolor="black",
)
plt.title("Preferred Values of C Across K=10 Folds", fontsize=10)
plt.xlabel("C Value", fontsize=10)
plt.ylabel("Frequency", fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

C = 0.01 performs in 5 out of 10 folds best. Thus, a more regularized model is prefered.

We can evaluate them once more separately a new model with the C hyperparameter set on the optimal value:

```python
log_standard_tuned = Pipeline(
    [
        ("preprocessor", StandardScaler()),
        (
            "classifier",
            LogisticRegression(max_iter=500, class_weight="balanced", C=0.01),
        ),
    ]
)
pipelines = [log_standard_tuned]

model_evaluation(pipelines, data_train, target_train)
```

### 9.4. <a id='toc9_4_'></a>[SVM with hyperparameter tuning](#toc0_)

#### 9.4.1. <a id='toc9_4_1_'></a>[Model selection](#toc0_)

As next we use the ***Support Vector Machines (SVM) classifier*** to train our data. It finds the hyperplane that maximizes the margin between the closest points of different classes. For non-linear data, it uses kernel functions to transform the data into higher dimensions. However, we need to select the appropriate kernel and tune the hyperparameters, which is a more complex task. Moreover, SVMs are less interpretable compared to simpler models like logistic regression.

#### 9.4.2. <a id='toc9_4_2_'></a>[Hyperparameter tuning and choice of the kernel](#toc0_)

There are two main hyperparameters to tune. The first is the choice of the **kernel**: A SVM is simpler with a ***linear kernel*** instead of ***radial basis function (RBF) kernel***. However, the latter one allows to catch non-linear relationships between the features. 

We start with the linear model and move then to a RBF kernel model. The RBF has the additional hyperparameter **gamma**, which we will tune then as well.

Moreover, there is a hinge loss with a L2 regularizer built in the SVM classifier. Therefore we need to tune the hyperparameter for the **regularization, C**. It is a tradeoff between a hard margin (very large C) and soft margin (small C), where the latter one is allowing inseparable class assignments.

The SVM does not work well with heterogeneous data types and needs scaling of the features, which we will achieve with the StandardScaler() as preprocessor in the pipeline.

```python
parameter_linear = {"classifier__C": np.logspace(-2, 3, 10)}
```

```python
parameter_rbf = {
    "classifier__C": np.logspace(-2, 2, 10),
    "classifier__gamma": np.logspace(-3, 1, 10),
}
```

#### 9.4.3. <a id='toc9_4_3_'></a>[Model training and hyperparameter C tuning](#toc0_)

We set up the pipeline first, which includes **Standardization** of the features:

```python
svm_linear_pipeline = Pipeline(
    [
        ("preprocessor", StandardScaler()),
        (
            "classifier",
            LinearSVC(class_weight="balanced"),
        ),
    ]
)
svm_linear_pipeline
```

We can see from the two boxplots above that the 20 folded cross-validations resulted in only 2 optimal values for C: 0.01 and 0.036. Plotted against the accuracy score, **C=0.01** shows a wider range but potentially also higher scores of **up to 0.85**. Therefore we want to use this hyperparameter for the final training cross-validation soon.

Moreover, the KNN classifier does not have a class_weight parameter. However, we can adjust sample weights manually within our pipeline:

```python
class WeightedPipeline(Pipeline):
    """Custom pipeline that passes sample weights to the classifier if supported."""
    def fit(self, X, y, **fit_params):
        sample_weights = compute_sample_weight(class_weight="balanced", y=y)
        if "classifier__sample_weight" in self.get_params():
            fit_params["classifier__sample_weight"] = sample_weights
        return super().fit(X, y, **fit_params)

knn_pipeline = WeightedPipeline(
    [
        ("preprocessor", MinMaxScaler()),
        ("classifier", knn),
    ]
)
knn_pipeline
```

### 9.6. <a id='toc9_6_'></a>[Random Forest model](#toc0_)

#### 9.6.1. <a id='toc9_6_1_'></a>[Model selection](#toc0_)

The random decision forests (**Random Forest**) model serves as the fourth model to test on our data. It is an ensemble learning method that works by creating a multitude of decision trees during training. Random Forests are *bagged* (bootstraped and aggregated) randomized decision trees. Randomized, because at each split only a random subset of features are selected. Each decision tree tends to overfit but averaging them reduces this threat. That makes Random Forest an excellent baseline model for binary feature datasets because it is robust and interpretable by using feature importances.

#### 9.6.2. <a id='toc9_6_2_'></a>[Hyperparameter tuning](#toc0_)

The main parameter to select in random forest is the **number of decision trees** `n_estimators` parameter. In general, the more trees in the forest, the better the generalization performance would be. However, adding trees slows down the fitting and prediction time. The goal is to balance computing time and generalization performance when setting the number of estimators. Tuning the n_estimators for random forests generally result in a waste of computer power. We just need to ensure that it is large enough so that doubling its value does not lead to a significant improvement of the validation error.The default value of n_estimators=100 should be save to go.


Instead, we can tune the hyperparameter `max_features`, which controls the **size of the random subset of features** to consider when looking for the best split when growing the trees: smaller values lead to more random trees with hopefully more uncorrelated prediction errors. However if it is too small, predictions can be too random, even after averaging with the trees in the ensemble.

We can also tune the different parameters that control the **depth of each tree** in the forest: `max_depth` (enforces growing deep symmetric trees) and `max_leaf_nodes` (restricts the growth of each tree). We want to tune both.
 
The hyperparameter `min_samples_leaf` controls the required **minimum number of samples at a leaf node**. This means that a split point (at any depth) is only done if it leaves at least min_samples_leaf training samples in each of the left and right branches. A small value means that some samples can become isolated when a tree is deep, promoting overfitting. A large value would prevent deep trees, which can lead to underfitting.

Please note that with random forest, trees are expected to be deep since we are seeking to overfit each tree on each bootstrap sample. Overfitting is mitigated when combining the trees altogether, whereas assembling underfitted trees (i.e. shallow trees) might also lead to an underfitted forest.

We want to test the following value space for the four discussed hyperparameters via `RandomizedSearchCV`:

*Note:* we iterate through this process multiple times, by changing the value space for each parameter based on the findings of the subsequent cross-validation tuning

```python
param_distributions = {
    "max_features": np.unique(np.logspace(0, 3, num=5, dtype=int)).tolist() + [None],
    "max_depth": np.unique(np.linspace(1, 10, num=6, dtype=int)).tolist() + [None],
    "max_leaf_nodes": np.unique(np.logspace(0, 3, num=5, dtype=int)).tolist() + [None],
    "min_samples_leaf": np.arange(1, 11).tolist(),
}
```

#### 9.6.3. <a id='toc9_6_3_'></a>[Model training](#toc0_)

```python
model_random_search = RandomizedSearchCV(
    RandomForestClassifier(class_weight="balanced", n_jobs=2),
    param_distributions=param_distributions,
    scoring="accuracy",
    n_iter=50,
    random_state=0,
    n_jobs=-1,
)

cv_results = cross_validate(
    model_random_search,
    data_train,
    target_train,
    cv=10,
    scoring="accuracy",
    return_estimator=True,
)
cv_results = pd.DataFrame(cv_results)

param_list = []
for estimator in cv_results["estimator"]:
    param_list.append(estimator.best_params_)

df_best_params = pd.DataFrame(param_list)
df_best_params["accuracy"] = cv_results["test_score"]
df_best_params.sort_values(by="accuracy", ascending=False, inplace=True)
df_best_params
```

From the table above it gets obvious, that the variance of accuracy of the best performing models is quite high. The test scores vary over multiple validation folds between 0.805 and 0.874.

However, the hyperparameter were not very stable across their values. Therefore, we want to improve our tuning process using **Bayesian Optimization**. This method is smarter than *random search* because it learns from previous evaluations and focuses on promising hyperparameters. We have to define hyperparameter space for the search method. Based on our results we refine it as following:

```python
search_space = {
    "max_features": [2, 5, 10, 50, 100, None],
    "max_depth": (4, 10),
    "max_leaf_nodes": (30, 1000),
    "min_samples_leaf": (1, 5),
}
```

Please note that in the code above **tuples** define a *continuous range* (if using Real) or a *discrete range* (if using Integer), while **lists** mean only these specific values will be tested.

We want to iterate through multiple iterations (`n_iter=50`) to find the best parameters but will not use a nested cross-validation due to the time-intensive procedure. Only the best model across the folds and its parameters are shown below:

```python
model_bayes_search = BayesSearchCV(
    RandomForestClassifier(class_weight="balanced", n_jobs=2, random_state=0),
    search_spaces=search_space,
    n_iter=100,
    scoring="accuracy",
    cv=20,
    random_state=0,
    n_jobs=-1,
)

model_bayes_search.fit(data_train, target_train)

best_params = model_bayes_search.best_params_
best_score = model_bayes_search.best_score_

df_best_params = pd.DataFrame([best_params])
df_best_params["accuracy"] = best_score

df_best_params
```

The test score reaches an accuracy of 0.83515, which is worst than finding a hyperparameter combination with `RandomSearchCV()`. Bayesian Optimization perhaps "over-shot" the perfect point where the hyperparameters optimize, because it also tries to optimize for overall robustness, avoiding overfitting to a specific set of hyperparameters.

## 10. <a id='toc10_'></a>[Suggestions for improvement](#toc0_)

There are multiple improvements possible for our project in solving this classification problem. Mainly we would like to **increase the Recall score**, as described in the conclusion (7.7) above. Hereinafter are four suggestions highlighted which could assist us in reaching that goal:

To improve further our **K-Nearest Neighbours model**, we could use more advanced ***hybrid approaches***. One would be to create feature subspaces, thus, separating the data into subsets based on feature types. We'd use different KNN models for each subset and ensemble their predictions. Regarding the distance metric we could experiment more with methods that takes into account the mixed data types (e.g. Gauf). Moreover, distance weighting would be interesting to test. It assigns weights to features based on their importance (e.g., higher weights for features more predictive of the outcome).

To improve further our **Random Forest model** we can turn to ***Gradient Boosting*** (e.g., XGBoost, LightGBM, or CatBoost). It is another ensemble technique that that builds models sequentially, each correcting the errors of its predecessor. It often performs better than random forests for classification problems but requires even more complex hyperparameter tuning and is computationally very expensive, thus requiring much more time.

Then we could make further use of **ensemble modeling**. Ideally, ensemble models should be obtained using different features or using algorithms of a different nature. In our case it would be very benefitial to set up a model that combines SVMs and Random Forest as classifiers, both performed very well as single classifiers in the models trained and tested.

To emphasize more on feature selection we could use techniques like ***SelectKBest*** or ***Recursive Feature Elimination (RFE)*** to select the most relevant features to include in the models. *SelectKBest* uses statistical tests to score and rank the features based on their relationship with the output variable and then selects the K features with the highest scores to be included in the final feature subset. *RFE* works by searching for a subset of features by starting with all features in the training dataset and successfully removing features until the desired number remains. This is achieved by fitting the given machine learning algorithm used in the core of the model, ranking features by importance, discarding the least important features, and re-fitting the model. This process is repeated until a specified number of features remains.
